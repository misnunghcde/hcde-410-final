## Explanation of the Code

#### 1. Loading the Datasets
The fertility rate and happiness index data are loaded from CSV files into pandas DataFrames for analysis. File paths must be accurately specified to ensure successful data loading.

#### 2. Renaming Columns for Consistency
The column `Country Name` in the fertility rate dataset is renamed to `Country name` for consistency, simplifying future merging operations.

#### 3. Reshaping the Fertility Data
The fertility data is reshaped from a wide format (years as columns) to a long format (rows for each year). This transformation facilitates filtering and merging by year.

#### 4. Cleaning and Converting Year Data
The `year` column is converted to a numeric format, and invalid or missing year values are removed. The `year` column is then cast to integers to ensure consistent data types.

#### 5. Standardizing Country Names
Alternative country names are replaced with standardized versions (e.g., `Korea, Rep.` → `South Korea`, `United States of America` → `United States`) to ensure consistency between datasets.In this part, Ai analyzed the data.

#### 6. Merging the Datasets
The fertility rate and happiness index datasets are merged on the common keys `Country name` and `year`. An inner join is used to include only rows with matching country names and years.

#### 7. Filtering Data by Year
The merged dataset is filtered to include only data from 2010 to 2020, focusing the analysis on recent trends.

#### 8. Saving the Processed Data
The cleaned and merged dataset is saved as `merged_data_2010_2020.csv`, ready for further analysis.

* This part was helped by AI.

```python
# Load the datasets
# Replace these file paths with the actual paths of your CSV files
fertility_data = pd.read_csv('fertility rate.csv')
happiness_data = pd.read_csv('Word Happiness Index.csv')

# Rename columns for consistency
fertility_data = fertility_data.rename(columns={"Country Name": "Country name"})
fertility_data_filtered = fertility_data.melt(id_vars=["Country name"], 
                                              var_name="year", 
                                              value_name="Fertility Rate")
fertility_data_filtered["year"] = pd.to_numeric(fertility_data_filtered["year"], errors="coerce")
fertility_data_filtered = fertility_data_filtered.dropna(subset=["year"])
fertility_data_filtered["year"] = fertility_data_filtered["year"].astype(int)

# Replace alternative country names to ensure consistency across both datasets (This part was helped by AI.)
country_name_mapping = {
    "Korea, Rep.": "South Korea",
    "United States of America": "United States",
    "Russian Federation": "Russia",
    "Czechia": "Czech Republic",
    "Egypt, Arab Rep.": "Egypt",
    "Iran, Islamic Rep.": "Iran",
    "Slovak Republic": "Slovakia",
    "Syrian Arab Republic": "Syria",
    "Venezuela, RB": "Venezuela",
    "Congo, Dem. Rep.": "Congo (Kinshasa)",
    "Congo, Rep.": "Congo (Brazzaville)",
    "Gambia, The": "Gambia",
    "Hong Kong SAR, China": "Hong Kong S.A.R. of China",
    "Cote d'Ivoire": "Ivory Coast",
    "Kyrgyz Republic": "Kyrgyzstan",
    "Lao PDR": "Laos",
    "Yemen, Rep.": "Yemen",
    "Korea, Dem. People's Rep.": "North Korea",
    "Turkiye": "Türkiye",
    "Viet Nam": "Vietnam"
}

# Apply country name mapping to both datasets
fertility_data_filtered["Country name"] = fertility_data_filtered["Country name"].replace(country_name_mapping)
happiness_data["Country name"] = happiness_data["Country name"].replace(country_name_mapping)

# Merge the datasets on "Country name" and "year"
merged_data = pd.merge(
    happiness_data,
    fertility_data_filtered,
    on=["Country name", "year"],
    how="inner"  # Use "inner" to ensure only matching data is included
)

# Filter for the years 2010 to 2020
merged_data_filtered = merged_data[(merged_data['year'] >= 2010) & (merged_data['year'] <= 2020)]

# Save the merged data to a CSV file
merged_data_filtered.to_csv('merged_data_2010_2020.csv', index=False)
```

### Fertility Rate Analysis by Quartile (2010-2020)

#### Background
The dataset contains fertility rate information for over 200 countries. To simplify visualization and analysis, the countries have been divided into quartiles based on their average fertility rates between 2010 and 2020. The quartiles were determined by calculating the average fertility rate for each country during this period and ranking them accordingly:
- **Q1 (Lowest 25%)**: Countries with the lowest fertility rates.
- **Q2 (25-50%)**: Countries in the second quartile.
- **Q3 (50-75%)**: Countries in the third quartile.
- **Q4 (Top 25%)**: Countries with the highest fertility rates.

This approach allows us to group countries and analyze trends within these quartiles. Additionally, South Korea's fertility rate trend is highlighted separately for comparison.

### Explanation of the Code

#### 1. Data Loading
The dataset `merged_data_2010_2020.csv` is loaded into a pandas DataFrame to analyze fertility rates and other relevant data from 2010 to 2020.

#### 2. Calculate the Average Fertility Rate by Country
The code calculates the average fertility rate for each country over the 2010-2020 period. This step helps summarize the fertility trends for each country into a single value for classification.

#### 3. Calculate the Quartiles
The quartiles are determined based on the average fertility rates:
- **Q1:** First quartile (lowest 25% of countries).
- **Q2:** Median fertility rate (25-50% of countries).
- **Q3:** Third quartile (50-75% of countries).
- **Q4:** Fourth quartile (75-100% of countries).

#### 4. Assign Quartiles to Each Country
Each country is assigned to one of the four quartiles based on its average fertility rate, categorizing countries into groups for analysis.

#### 5. Merge Quartile Information with the Main Dataset
The quartile information is added to the original dataset. This allows trends for each quartile to be analyzed over time.

#### 6. Plot the Average Fertility Rate Trends by Quartile
The average fertility rate for each quartile is calculated for each year (2010-2020), and trends are plotted as separate lines for Q1 to Q4.

#### 7. Highlight South Korea's Fertility Rate Trend
South Korea's fertility rate is highlighted on the chart with a distinctive dashed black line for comparison against quartiles.

#### 8. Customize and Display the Chart
The chart is enhanced with titles, labels, legends, and grid lines for better readability. It shows fertility rate trends for all quartiles and compares them to South Korea's unique trend.


```python
# data load
data = pd.read_csv('merged_data_2010_2020.csv')

# calute a TRT for the given period
avg_fertility_by_country = (
    data.groupby('Country name')['Fertility Rate']
    .mean()
    .reset_index()
    .rename(columns={'Fertility Rate': 'Avg Fertility Rate'})
)

# caculate the quartile
q1 = avg_fertility_by_country['Avg Fertility Rate'].quantile(0.25)
q2 = avg_fertility_by_country['Avg Fertility Rate'].quantile(0.5)
q3 = avg_fertility_by_country['Avg Fertility Rate'].quantile(0.75)

# assign quartile
def assign_quartile(value):
    if value <= q1:
        return 'Q1 (Lowest 25%)'
    elif value <= q2:
        return 'Q2 (25-50%)'
    elif value <= q3:
        return 'Q3 (50-75%)'
    else:
        return 'Q4 (Top 25%)'

# add quartile
avg_fertility_by_country['Quartile'] = avg_fertility_by_country['Avg Fertility Rate'].apply(assign_quartile)

# merge quartile data to data
data_with_quartiles = pd.merge(
    data,
    avg_fertility_by_country[['Country name', 'Quartile']],
    on='Country name'
)

# plot the average of quartile
plt.figure(figsize=(14, 8))

for quartile, group in data_with_quartiles.groupby('Quartile'):
    # caculate average of quartiles
    quartile_mean = group.groupby('year')['Fertility Rate'].mean()
    plt.plot(
        quartile_mean.index,
        quartile_mean.values,
        label=f'{quartile} (Average)',
        linewidth=2
    )

# plot south korea data
korea_data = data[data['Country name'] == 'South Korea']
if not korea_data.empty:
    plt.plot(
        korea_data['year'],
        korea_data['Fertility Rate'],
        label='South Korea',
        linewidth=3,
        linestyle='--',
        color='black'
    )

plt.title('Average Fertility Rate Trends by Quartile (2010-2020) with South Korea', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Fertility Rate', fontsize=12)
plt.xticks(data['year'].unique(), fontsize=10)
plt.legend(title='Quartile', fontsize=10, loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

## Fertility Rate Correlation Analysis Report

### Research Background
The analysis focuses on understanding the relationship between fertility rates and various socioeconomic and demographic indicators. To measure these relationships, the **Pearson correlation coefficient** is employed:
- **Pearson Correlation Coefficient**: A statistical metric quantifying the strength and direction of linear relationships between continuous variables. Its range includes:
 - **-1**: Perfect negative correlation (inverse relationship between variables)
 - **0**: No correlation
 - **1**: Perfect positive correlation (direct proportional relationship)
- This analysis highlights indicators most strongly associated with fertility rates.

### Explanation of the Analytical Approach

#### 1. Data Loading
The dataset `merged_data_2010_2020.csv` is loaded into a panda.

#### 2. Numeric Column Identification
Utilizing pandas' `select_dtypes` function, the code methodically identifies all numeric columns within the dataset.

#### 3. Fertility Rate Column Verification
A critical validation step confirms the presence of the Fertility Rate column among the identified numeric columns.

#### 4. Correlation Calculation
Upon confirming the Fertility Rate column's existence:
- Pearson correlation coefficients are calculated between the Fertility Rate and all other numeric columns using the pandas' `corr` method
- Correlations are systematically sorted in descending order to prioritize the most significant relationships
- Results are meticulously stored in a data frame for comprehensive visualization and interpretation

#### 5. Data Presentation
Calculated correlations provide an immediate overview of the relationships between Fertility Rate and various indicators.

#### 6. Visualization Strategy
A bar chart is constructed to visually represent correlation strengths:
- Each bar represents a numeric indicator's correlation with the Fertility Rate
- Color differentiation highlights positive and negative correlations


```python
# Load the dataset
file_path = "merged_data_2010_2020.csv"
data = pd.read_csv(file_path)

# Identify numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
fertility_column = 'Fertility Rate'

# Ensure 'Fertility Rate' is in the dataset
if fertility_column in numeric_columns:
    # Calculate correlations
    correlations = data[numeric_columns].corr()[fertility_column].drop(fertility_column)
    correlation_df = correlations.sort_values(ascending=False).to_frame(name='Correlation with Fertility Rate')
    
    # Display the correlation data
    print("Correlations with Fertility Rate:")
    print(correlation_df)

    # Plot the correlations
    plt.figure(figsize=(10, 6))
    correlation_df.plot(kind='bar', legend=False, color='skyblue', figsize=(10, 6))
    plt.title("Correlations with Fertility Rate (2010-2020)")
    plt.ylabel("Correlation")
    plt.xlabel("Indicators")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print(f"'{fertility_column}' is not in the dataset.")
    
# This code block was helped by AI.
```

## Code Explanation: Comparing Global vs South Korean Correlations

#### Purpose of the Analysis
The primary objective of this analysis is to identify and compare the key factors influencing fertility rates on a global scale versus those specific to South Korea. By examining these correlations, we aim to uncover unique patterns or relationships that are particularly relevant to South Korea's socioeconomic and cultural context.

#### 1. Verify Data Contains Fertility Rate Column
The code first checks if the `Fertility Rate` column exists in the numeric columns of the dataset. With this column, the analysis can proceed.

#### 2. Exclude Irrelevant Columns
The `year` column is temporal and is excluded from the numeric columns to ensure meaningful correlations with the fertility rate.

#### 3. Calculate Global Correlations
The Pearson correlation coefficients for all numeric indicators against the fertility rate are calculated globally (across all countries).

#### 4. Focus on South Korea
For South Korea precisely, the same correlations are calculated using only the subset of data where the `Country name` is "South Korea." This allows for comparing the relationships specific to South Korea against global trends.

#### 5. Merge Results for Comparison
The global and South Korea-specific correlations are merged into a single data frame to enable side-by-side comparison.

#### 6. Visualize Comparison
A side-by-side bar graph is plotted to compare the correlations of different indicators with fertility rates globally and within South Korea:
   - The x-axis represents the indicators.
   - The y-axis shows the correlation coefficients.
   - Different bars are used to distinguish between global and South Korea-specific correlations.

This visualization highlights how South Korea's unique socioeconomic and cultural context may influence it.

```python
# Ensure 'Fertility Rate' is in the dataset
if fertility_column in numeric_columns:
    # Remove 'year' from numeric columns
    numeric_columns = numeric_columns.drop('year', errors='ignore')
    
    # Calculate global correlations
    correlations = data[numeric_columns].corr()[fertility_column].drop(fertility_column)
    correlation_df_filtered = correlations.sort_values(ascending=False).to_frame(name='Global Correlation')
    
    # Filter data for South Korea and calculate correlations
    country_name = "South Korea"
    south_korea_data = data[data["Country name"] == country_name]
    south_korea_correlations = south_korea_data[numeric_columns].corr()[fertility_column].drop(fertility_column)
    south_korea_correlation_df = south_korea_correlations.sort_values(ascending=False).to_frame(name='South Korea Correlation')
    
    # Merge global and South Korea correlations for comparison
    comparison_df = pd.concat([correlation_df_filtered, south_korea_correlation_df], axis=1)
    
    # Plot side-by-side bar graph
    plt.figure(figsize=(14, 7))
    comparison_df.plot(kind='bar', figsize=(14, 7), width=0.8)
    plt.title('Comparison of Correlations with Fertility Rate (Global vs South Korea)', fontsize=16)
    plt.ylabel('Correlation', fontsize=12)
    plt.xlabel('Indicators', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title='Dataset', fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print(f"'{fertility_column}' is not in the dataset.")
```

### Code Explanation: Analysis of Births Outside Marriage and Fertility Rate (2010-2020)

#### Purpose of the Code
This code merges and analyzes two datasets:
1. **Percentage of births outside marriage** for each country from 2010 to 2020.
2. **Fertility rates** for the same period.

The goal is to visualize the relationship between these two variables across countries, highlighting potential patterns or correlations.

#### Code Breakdown

1. **Data Loading**:
   - Load two datasets:
     - Births outside marriage data (`births_outside_marriage.csv`).
     - Fertility rate data (`merged_data_2010_2020.csv`).

2. **Filter Data for Relevant Years**:
   - Select columns corresponding to the years 2010 to 2020 from both datasets.
   - Retain only rows where at least one value in the selected years is not `NaN` to ensure meaningful analysis.

3. **Calculate Averages**:
   - For each country, compute:
     - The average percentage of births outside marriage (2010-2020).
     - The average fertility rate (2010-2020).
   - These averages summarize the data for visualization and comparison.

4. **Standardize and Map Country Names**:
   - Align country names between the two datasets using a mapping dictionary to ensure successful merging.
  
5. **Merge Datasets**:
   - Combine the two datasets into a single data frame using an inner join on the standardized country names.
   - This allows for side-by-side analysis of births outside marriage and fertility rates.

6. **Sort and Prepare Data for Visualization**:
   - Sort the merged dataset by average fertility rate in descending order to highlight countries with the highest fertility rates.

7. **Plot Dual-Axis Chart**:
   - **Bar Chart (Primary Y-Axis)**:
     - Represents the average percentage of births outside marriage for each country.
     - Countries are ordered along the x-axis.
   - **Line Chart (Secondary Y-Axis)**:
     - Displays the average fertility rate for each country.
     - This enables a direct comparison of the two metrics.

8. **Customize Visualization**:
   - Titles, axis labels, and grid lines are added for clarity.

#### Insights from the Visualization
- Countries with higher percentages of births outside marriage may exhibit distinct fertility trends compared to those with lower percentages.
The dual-axis chart clearly observes relationships or contrasts between non-marital births and overall fertility rates.


```python
# Load the datasets
file_path_births = "births_outside_marriage.csv"  # Path to births data
file_path_fertility = "merged_data_2010_2020.csv"  # Path to fertility data

births_data = pd.read_csv(file_path_births)
fertility_data = pd.read_csv(file_path_fertility)

# Clean and preprocess the births data
births_data.replace('..', pd.NA, inplace=True)
births_data = births_data.apply(pd.to_numeric, errors='ignore')

# Select data from 2010 to 2020 for births
years = [str(year) for year in range(2010, 2021)]
births_data_filtered = births_data[['Country'] + years].copy()

# Drop rows where all values for 2010-2020 are NaN
births_data_filtered = births_data_filtered.dropna(subset=years, how='all')

# Calculate the average percentage of births outside marriage for each country
births_data_filtered['Average (2010-2020)'] = births_data_filtered[years].mean(axis=1, skipna=True)

# Standardize the 'Country' column
births_data_filtered['Country'] = births_data_filtered['Country'].astype(str)
births_data_filtered.rename(columns={'Country': 'Country Name'}, inplace=True)

# Clean and preprocess the fertility data
fertility_data.replace('..', pd.NA, inplace=True)
fertility_data = fertility_data.apply(pd.to_numeric, errors='ignore')

# Select and pivot fertility data for relevant years
fertility_data_filtered = fertility_data[['Country name', 'year', 'Fertility Rate']]
fertility_data_pivoted = fertility_data_filtered.pivot_table(
    index='Country name', 
    columns='year', 
    values='Fertility Rate'
).reset_index()

# Align fertility data column names
fertility_data_pivoted.columns = ['Country Name'] + years
fertility_data_pivoted['Average Fertility Rate (2010-2020)'] = fertility_data_pivoted[years].mean(axis=1, skipna=True)

# Standardize country names for merging (AI helped this code)
country_name_mapping = {
    "Korea": "South Korea",
    "Korea, Rep.": "South Korea",
    "Czech Republic": "Czechia",
    "Türkiye": "Turkey"
}
births_data_filtered['Country Name'] = births_data_filtered['Country Name'].replace(country_name_mapping)
fertility_data_pivoted['Country Name'] = fertility_data_pivoted['Country Name'].replace(country_name_mapping)

# Merge the datasets
merged_data = pd.merge(
    births_data_filtered,
    fertility_data_pivoted[['Country Name', 'Average Fertility Rate (2010-2020)']],
    on='Country Name',
    how='inner'
)

# Sort the merged data by fertility rate
merged_data_sorted = merged_data.sort_values(by='Average Fertility Rate (2010-2020)', ascending=False)

# Plot the dual-axis chart
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar plot for births outside marriage
color = 'skyblue'
ax1.bar(
    merged_data_sorted['Country Name'], 
    merged_data_sorted['Average (2010-2020)'], 
    color=color, 
    alpha=0.7, 
    label='Births Outside Marriage (%)'
)
ax1.set_xlabel('Country', fontsize=12)
ax1.set_ylabel('Births Outside Marriage (%)', color=color, fontsize=12)
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x', rotation=90, labelsize=10)
ax1.set_title('Average Births Outside Marriage and Fertility Rate (2010-2020)', fontsize=16)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Line plot for fertility rates using a secondary y-axis
ax2 = ax1.twinx()
color = 'red'
ax2.plot(
    merged_data_sorted['Country Name'], 
    merged_data_sorted['Average Fertility Rate (2010-2020)'], 
    color=color, 
    marker='o', 
    label='Fertility Rate'
)
ax2.set_ylabel('Fertility Rate', color=color, fontsize=12)
ax2.tick_params(axis='y', labelcolor=color)

# Adjust layout
fig.tight_layout()
plt.show()
```
